```python
import numpy as np
from sklearn.preprocessing import KBinsDiscretizer
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\Nitro\\Desktop\\Inteligencia_Artificial\\PreProcessamento_Udemy\\sample_dataset.csv").iloc[:, 0:5].dropna()
df
```

Bins with uniform width

```python
binner = KBinsDiscretizer(strategy='uniform', n_bins=5)
binner.fit_transform(df)
```

```python
binner = KBinsDiscretizer(strategy='uniform', n_bins=5, encode = 'onehot-dense')
binner.fit_transform(df)
```

```python
binner = KBinsDiscretizer(strategy='uniform', n_bins=5, encode = 'ordinal')
binner.fit_transform(df)
```

Binning using quantiles

```python
binner = KBinsDiscretizer(strategy='quantile', n_bins=5, encode = 'ordinal')
binner.fit_transform(df)
```

```python
np.unique(binner.fit_transform(df)[:,0], return_counts=True) # Retorna o nÃºmero de cada elemento na lista
```

Binning using K-Means

```python
binner = KBinsDiscretizer(strategy='kmeans', n_bins=3, encode = 'ordinal')
binner.fit_transform(df)
```
