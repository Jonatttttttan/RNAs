```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer, make_column_selector
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv("C:\\Users\\Nitro\\Desktop\\Inteligencia_Artificial\\PreProcessamento_Udemy\\sample_dataset.csv")
```

```python
transformer = ColumnTransformer([
    ('numerical', Pipeline([
        ('cleaner', SimpleImputer(strategy='median')), # Adiciona dados faltantes
        ('scaler', MinMaxScaler()) # Normaliza
    ]), make_column_selector(dtype_exclude='object'))
])

transformer.fit_transform(df)
```

```python
transformer.set_params(numerical__scaler = StandardScaler())
```
