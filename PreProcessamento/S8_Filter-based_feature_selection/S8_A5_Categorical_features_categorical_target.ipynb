```python
import pandas as pd
from sklearn.feature_selection import SelectKBest, mutual_info_classif, chi2
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder
```

```python
df = pd.read_csv("C:\\Users\\Nitro\\Desktop\\Inteligencia_Artificial\\PreProcessamento_Udemy\\sample_dataset_bins.csv")
```

```python
df
```

```python
df['radius error'] =  pd.cut(df['radius error'], 10)
```

```python
df
```

```python
X = df.iloc[:, 0:10]
y = df.iloc[:, -1]
```

```python
X = OrdinalEncoder().fit_transform(X)
y = LabelEncoder().fit_transform(y)
```

```python
X
```

```python
y
```

Mutual information

```python
selector = SelectKBest(lambda X, y: mutual_info_classif(X, y, discrete_features=True), k = 5)
```

```python
selector.fit_transform(X,y)
```

Chi-square

```python
selector = SelectKBest(chi2, k = 5)
```

```python
selector.fit_transform(X,y)
```
