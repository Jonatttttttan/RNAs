```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE, SMOTENC
```

```python
df = pd.read_csv("C:\\Users\\Nitro\\Desktop\\Inteligencia_Artificial\\PreProcessamento_Udemy\\sample_dataset.csv").iloc[:,-10:].dropna()
df
```

```python
X = df.iloc[:,0:-1]
y = df['target']
```

```python
X_scaled = StandardScaler().fit_transform(X)
```

```python
y.value_counts()
```

SMOTE with numerical features

```python
resampler = SMOTE(random_state = 0)
```

```python
X_res, y_res = resampler.fit_resample(X_scaled, y)
```

```python
y_res.value_counts()
```

SMOTE with categorical features

```python
df = pd.read_csv("C:\\Users\\Nitro\\Desktop\\Inteligencia_Artificial\\PreProcessamento_Udemy\\sample_dataset.csv").loc[:,['mean radius', 'mean texture', 'mean perimeter', 'area error','target']].dropna()
```

```python
df
```

```python
X = df.iloc[:,0:-1]
y = df['target']
```

```python
resampler = SMOTENC(random_state=0, categorical_features=[3])
```

```python
X_res, y_res = resampler.fit_resample(X,y)
```

```python
y_res.value_counts()
```
