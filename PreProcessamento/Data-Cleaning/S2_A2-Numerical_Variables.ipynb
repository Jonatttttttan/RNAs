```python
import pandas as pd
from sklearn.impute import SimpleImputer
df = pd.read_csv("C:\\Users\\Nitro\\Desktop\\Inteligencia_Artificial\\PreProcessamento_Udemy\\sample_dataset.csv")
```

```python
X = df.iloc[:,0:3] # Todas as linhas e as 3 primeiras colunas
```

```python
X
```

Replace blanks with the mean value

```python
cleaner = SimpleImputer(strategy='mean')
```

```python
cleaner.fit(X)
```

```python
cleaner.transform(X)
```

Replave blanks with median value

```python
cleaner = SimpleImputer(strategy='median')
cleaner.fit_transform(X)
```

Replace blanks with fixed value

```python
cleaner = SimpleImputer(strategy = 'constant', fill_value = 0)
cleaner.fit_transform(X)
```
